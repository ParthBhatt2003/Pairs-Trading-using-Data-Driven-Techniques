```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
import math
```

```python
fig = plt.figure(1, figsize = (8, 8))
delta = 0.025
T, t = 1, 0.5        # change value of t to get different results
sig, r, K = 0.3, 0.12, 50
x, y = np.meshgrid(np.arange(1, 151, delta), np.arange(1, 151, delta))
d1 = ( np.log(x/K) + (r + sig*sig/2)*(T-t))/(sig*np.sqrt(T-t))
d2 = d1 - sig*np.sqrt(T-t)
call = y - scipy.stats.norm.cdf(d1)*x + scipy.stats.norm.cdf(d2)*K/pow(np.e, (r*(T-t)))
plt.contour(x, y, call, [0])
plt.text(20, 80, f"t = {t}", fontsize = 22)
plt.xlabel("$Stock price", fontsize = 20)
plt.ylabel("$call option price/Premium", fontsize = 20)

plt.show()
```
